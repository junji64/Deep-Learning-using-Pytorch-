# Matplotlib

**Matplotlib**은 플랫폼에 상관없이 다양한 하드 카피 형식과 대화식 환경으로 출판 품질의 그림을 생성하는 Python 2D 플로팅 라이브러리입니다. **matplotlib**는 python 스크립트, python 및 ipython 쉘, 웹 응용 프로그램 서버 및 6 개의 그래픽 사용자 인터페이스 툴킷에서 사용할 수 있습니다.

**Matplotlib**은 쉬운 일을 쉽게, 어려운 일은 가능하게 만들려고 합니다. 단지 몇 줄의 코드로 플롯, 히스토그램, 파워 스펙트럼, 막대 차트, 오류 차트, 스캐터플롯 등을 생성 할 수 있습니다.

라이브러리 설명서 : http://matplotlib.org/

```python
import matplotlib.pyplot as plt
```

```python
x = [ 1, 2, 3]
y = [ 5, 7, 4]

plt.plot(x, y)
plt.xlabel('x axis')
plt.ylabel('Values')
plt.title('Test Graph')

plt.show()
```

```python
x = [ 1, 2, 3]
y = [ 5, 7, 4]

x2 = [1, 2, 3]
y2 = [10, 14, 12]

plt.plot(x, y, label='First Line')
plt.plot(x2, y2, label='Second Line')
plt.xlabel('x axis')
plt.ylabel('Values')
plt.title('Test Graph')
plt.legend()
plt.show()
```

```python
import numpy as np

x = np.linspace(0,5,10)
y = x ** 2
plt.plot(x, y, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.show()
```

```python
x = [2, 4, 6, 8, 10]
y = [6, 7, 8, 2, 4]

x2 = [1, 3, 5, 7, 9]
y2 = [7, 8, 2, 4, 2]

plt.bar(x, y, label="Bars 1", color="r")
plt.bar(x2, y2, label="Bars 2", color="g")

plt.show()
```

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x, y, label="scatter plot", color="b", marker="*")
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=3, figsize=(12, 4))

x = np.linspace(0, 5, 10)

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].grid()

axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_xlim([2,5])
axes[2].set_ylim([0,60])
axes[2].set_title("custom axis range")

plt.show()
```

```python
# example data
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(437)

num_bins = 50

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=1)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '--')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```
